```python
import pandas as pd
from datetime import datetime
import plotly.offline as py
# from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

py.init_notebook_mode(connected=True)
import plotly.graph_objs as go

df = pd.read_csv('adx.csv')

data = [go.Scatter(x=df.Date, y=df['ADX'], name='ADX')]

py.iplot(data, filename = 'time-series-simple')
```

```python
import pandas as pd
from datetime import datetime

df = pd.read_csv('adx.csv')

ADX = go.Scatter(x=df.Date, y=df['ADX'], name='ADX')
High = go.Scatter(x=df.Date, y=df['High'], name='High')
data = [ADX,High]

py.iplot(data, filename = 'time-series-simple')
```

```python
df = pd.read_csv("RELIANCE.csv")

trace_high = go.Scatter(
    x=df.TIMESTAMP,
    y=df['CLOSE'],
    name = "Close",
    line = dict(color = '#17BECF'),
    opacity = 0.8)

trace_low = go.Scatter(
    x=df.TIMESTAMP,
    y=df['HIGH'],
    name = "High",
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)

data = [trace_high,trace_low]

layout = dict(
    title='Time Series with Rangeslider',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Time Series with Rangeslider")
# py.iplot(data)
```

```python
import plotly.offline as py
import plotly.graph_objs as go

import pandas as pd

df = pd.read_csv("RELIANCE.csv")

trace_high = go.Scatter(
                x=df.TIMESTAMP,
                y=df['HIGH'],
                name = "High",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

trace_low = go.Scatter(
                x=df.TIMESTAMP,
                y=df['LOW'],
                name = "Low",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)

data = [trace_high,trace_low]

trace_close = go.Scatter(
                x=df.TIMESTAMP,
                y=df['CLOSE'],
                name = "Close",
                line = dict(color = '#CD5C5C'),
                opacity = 0.8)

data = [trace_high,trace_low]

trace_open = go.Scatter(
                x=df.TIMESTAMP,
                y=df['OPEN'],
                name = "Open",
                line = dict(color = '#228B22'),
                opacity = 0.8)

data = [trace_high,trace_low, trace_open, trace_close]

layout = go.Layout(
    title=go.layout.Title(
        text='RELIANCE',
        xref='paper',
        x=0
    )
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "RELIANCE")
```
